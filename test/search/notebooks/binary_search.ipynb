```python
from py_puzzle.search.algorithm import BinarySearch

binary_search = BinarySearch(arr=[1,2,3,4,5,6,6,6,6,7,8,9,10], target=6).first_value()
print(binary_search)
assert binary_search["Output"] == 5
```

```python
bs = BinarySearch()
tgt_value = bs.input.index(bs.target)
result = bs.process()
print(result, "ExpectedOutput:", tgt_value)

try:
    assert result["Output"] == tgt_value
except AssertionError:
    result = bs.first_value()
    print("Retrying with first_value:", result)
    assert result["Output"] == tgt_value
```
